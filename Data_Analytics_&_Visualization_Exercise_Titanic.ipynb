## Applying Apyori Algorithm to Find Association for the Variables

```python
#Collecting the inferred rules in the dataframe
association_rules = ap(Titanic_df.values, min_support = 0.005, min_confidence = 0.8, min_length = 2)
association_results = list(association_rules)

filteredResults = []

#Filtering our results to justify rules that the rhs is survived
for result in association_results:
    for entry in result.ordered_statistics:
        if entry.items_add == frozenset({'Yes'}):
            filteredResults.append(entry)
print("\nNumber of rules: {0}\n".format(len(filteredResults)))
```

```python
sortedResults = sorted(filteredResults, key= lambda x: x.lift, reverse = True)
for results in sortedResults:
    print(str(results))
```
