## Encode the target

```python
inputs = Adults_pd.drop('class', axis='columns')

target = Adults_pd['class']
```

```python
Adults_pd['class'].replace('<=50K', 0, inplace=True)
Adults_pd['class'].replace('>50K', 1, inplace=True)
```

### Encoding

```python
le_workclass = LabelEncoder()

le_education = LabelEncoder()

le_maritalstatus = LabelEncoder()

le_occupation = LabelEncoder()

le_relationship = LabelEncoder()

le_race = LabelEncoder()

le_sex = LabelEncoder()

le_capitalgain = LabelEncoder()
```

```python
#create columns in dataframe

#Encode dataframe to be compatible with decision tree

inputs['workclass_n'] = le_workclass.fit_transform(inputs['workclass'])

inputs['education_n'] = le_education.fit_transform(inputs['education'])

inputs['maritalstatus_n'] = le_maritalstatus.fit_transform(inputs['maritalstatus'])

inputs['occupation_n'] = le_occupation.fit_transform(inputs['occupation'])

inputs['relationship_n'] = le_relationship.fit_transform(inputs['relationship'])

inputs['race_n'] = le_race.fit_transform(inputs['race'])

inputs['sex_n'] = le_sex.fit_transform(inputs['sex'])

inputs['capitalgain_n'] = le_capitalgain.fit_transform(inputs['capitalgain'])

print(inputs['workclass'])
print(inputs['workclass_n'])

print(inputs.head())
```

```python
#drop extra columns

inputs_n = inputs.drop(['workclass','education','maritalstatus','occupation', 'relationship','race','sex','capitalgain'], axis='columns')

print(inputs_n.head())
```

```python
x = inputs_n

y = target

rs = 42
```

```python
## Train_Test_Split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state=rs)
```

## Encode Dataframe for Decision Tree

```python
conditions = ["age", "fnlwgt", "educationnum", "capitalloss", "hoursperweek"]
```

```python
for col in conditions:
    dummies = pd.get_dummies(Adults_pd[col], prefix=col)
    Adults_pd = pd.concat([Adults_pd, dummies], axis=1)
Adults_pd.drop(conditions, axis=1, inplace=True)

# Print the updated DataFrame
print(Adults_pd.head())
print(Adults_pd.tail())
```

```python
classifiers = [gnb, dtc]
classifier_names = ["Gaussian NB", "Decision Tree"]
```

```python
## Fit the model Decision Tree

model = dtc.fit(x_train, y_train)
```

```python
## Predict

y_pred = model.predict(x_test)
```

```python
## Generating Confusion Matrix

results = metrics.confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", results)
```

```python
## Predict

y_pred = model.predict(x_test)
```

```python
## Evaluating the model

print("Accuracy:\n", metrics.classification_report(y_test, y_pred))
```
