## Part 1 Data Categorization and Cleaning

```python
## Checking duplicates
data_df.duplicated()
```

```python
##Dropping duplicates
data2_df = data_df.copy()
data2_df.drop_duplicates(inplace=True)
data2_df
```

```python
## Identifying numeric columns
def get_dtype_columns(df, type_):
    if type_ == 'numeric':
        return df.select_dtypes(include=[np.number]).columns.values
    else:
        return df.select_dtypes(exclude=[np.number]).columns.values

numeric_cols = get_dtype_columns(data2_df, 'numeric')
print(numeric_cols)
```

```python
## Identifying non-numeric columns
def get_dtype_columns(df, type_):
    if type_ == 'non-numeric':
        return df.select_dtypes(exclude=[np.number]).columns.values
    else:
        return df.select_dtypes(include=[np.number]).columns.values

non_numeric_cols = get_dtype_columns(data2_df, 'non-numeric')
print(non_numeric_cols)
```

```python
### Percentage of missing values
for col in data2_df.columns:
    percent_of_missing = np.mean(data2_df[col].isna())
    print('{} - {}%'.format(col, round(percent_of_missing*100)))
```

```python
## Numeric features with less than 2% missing values
for col in numeric_cols:
    if (np.mean(data2_df[col].isnull()*100)) < 2:
        data2_df[col] = data2_df[col].fillna(data2_df[col].median())

data2_df
```

```python
## Numeric feature that has missing values 5% or more
for col in numeric_cols:
    if (np.mean(data2_df[col].isnull()*100)) >= 5:
        data2_df[col] = data2_df[col].fillna(data2_df[col].mode()[0])

data2_df
```

```python
data2_df['price'].isna().sum()
```

```python
## Replacing all non-numeric feature with less than 5% missing values with "_NA_"
non_numeric_cols = get_dtype_columns (data2_df, "non-numeric")
for num_col in non_numeric_cols:
  if (np.mean(data2_df [num_col].isnull()*100)) < 5:
    data2_df[num_col] = data2_df[num_col].fillna("_NA_")

data2_df
```

```python
### Percentage of missing values
for col in data2_df.columns:
    percent_of_missing = np.mean(data2_df[col].isna())
    print('{} - {}%'.format(col, round(percent_of_missing*100)))
```

```python
data2_df['sold_date'].mode()
```

```python
## Replacing non-numeric feature that has 5% or more missing values with the most frequent value for this feature.
non_numeric_cols = get_dtype_columns (data2_df, "non-numeric")
for num_col in non_numeric_cols:
  if (np.mean(data2_df [num_col].isnull()*100)) >= 5:
    data2_df[num_col] = data2_df[num_col].fillna(data2_df[num_col].mode()[0])

data2_df
```

```python
### Percentage of missing values
for col in data2_df.columns:
    percent_of_missing = np.mean(data2_df[col].isna())
    print('{} - {}%'.format(col, round(percent_of_missing*100)))
```

```python
## Replace any remaining missing numeric value with 0
for col in get_dtype_columns(data2_df, 'numeric'):
    data2_df[col] = data2_df[col].fillna('0')
data2_df
```

```python
## Replacing any remaining missing non-numeric value with the word '_MISSING_'
for col in get_dtype_columns(data2_df, 'non-numeric'):
    data2_df[col] = data2_df[col].fillna('_MISSING_')
data2_df
```

```python
data2_df.shape
```

```python
## For each numeric feature in the dataset, drop observations of this feature that are greater than (mean + 3 * std) or less than (mean – 3 * std).
## Where the mean is the mean of the data that is not missing and std is the standard deviation of the data that is not missing of each feature

for col in numeric_cols:
    mean_cal = np.mean(data2_df[col])
    standard_dev = np.std(data2_df[col])
    first_order = mean_cal + 3*standard_dev
    second_order = mean_cal - 3*standard_dev

    data2_df.drop(data2_df[(data2_df[col] > first_order) | (data2_df[col] < second_order)].index, inplace=True)

data2_df
```
